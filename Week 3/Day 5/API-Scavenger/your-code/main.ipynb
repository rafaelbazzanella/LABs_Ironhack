## Challenge 1: Fork Languages

You will find out how many programming languages are used among all the forks created from the main lab repo of your bootcamp.

`````python
# import libraries here
import json
import requests
import pandas as pd
from tqdm.auto import tqdm
from pandas.io.json import json_normalize;
`````

Assuming the main lab repo is ironhack-datalabs/mad-oct-2018, you will:

#### 1. Obtain the full list of forks created from the main lab repo via Github API.

To list forks, we can use the GET method. As explained in the GitHub API documentation, we need to make the request to: GET /repos/:owner/:repo/forks.

`````python
# your code here
response = requests.get('https://api.github.com/repos/ironhack-datalabs/mad-oct-2018/forks')
df = response.json()
`````

`````python
forks = json_normalize(df)
forks.head()
`````

#### 2. Loop the JSON response to find out the language attribute of each fork. Use an array to store the language attributes of each fork.
Hint: Each language should appear only once in your array.
Print the language array. It should be something like: ["Python", "Jupyter Notebook", "HTML"]

`````python
# your code here
language = list(forks['language'].unique())
`````

`````python
language
`````

`````python
# Looping
languages = [];

for language in forks['language']:
    if language not in languages:
        languages.append(language); 
        
print(languages);
`````

`````python
# your code here
url = 'https://api.github.com/repos/ironhack-datalabs/scavenger/contents'
response3 = requests.get(url)
contents = pd.json_normalize(response3.json())
folder_names = contents['name']

`````

`````python
contents.head()
`````

`````python
folder_names
`````

`````python
#cada link que vamos usar 
link=[]
for name in folder_names[1:]:
    link.append(url+'/'+ name)
    
`````

`````python
lst_scavenger=[]
for each_link in tqdm(link):
    response_each = requests.get(each_link)
    content_each = pd.json_normalize(response_each.json())
    if content_each['name'][0].startswith('.'):
        scavenger = content_each['name']
        lst_scavenger.append(each_link+'/'+scavenger)
`````

`````python
lst_scavenger
`````

`````python
l = [item for sublist in [list(item) for item in lst_scavenger] for item in sublist]
`````
